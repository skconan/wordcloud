```python
import matplotlib.pyplot as plt
import pandas as pd
df =  pd.read_csv('A_Series_of_Unfortunate_Events_sentimental.csv')

data = 'Neutral', 'Positive','Negative'

neu = df.loc[1][1]
pos = df.loc[2][1]
neg = df.loc[3][1]
value = [neu, pos, neg] 
colors = ['#FEFCAD','#A0E8AF','#F0544F']

plt.pie(value, labels=data, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')
plt.title("A Series of Unfortunate Events")
plt.show()
```
